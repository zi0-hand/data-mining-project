```python
# '1인가득밥상-강남점'을 기준으로 메뉴 살펴보기

df_menu_num = df[df['Restaurant']=='1인가득밥상-강남점']
df_menu_cont = df_menu_num['Menu'].value_counts().to_frame()
df_menu_cont[df_menu_cont['Menu']>30]
```

```python
# 1개의 메뉴만을 시킨 리뷰 추출하기 (/ 하나당 메뉴하나니까)

df_menu_num['menu_count'] = df_menu_num.Menu.str.count('/')
```

```python
df_menu_num_1 = df_menu_num[df_menu_num['menu_count']==1]
#df_menu_num_1['Menu'].value_counts()

df_menu_count = df_menu_num_1['Menu'].value_counts().to_frame()
df_menu_count[df_menu_count['Menu']>2]
```

```python
df_menu_num_1.columns
```

```python
col = ['Restaurant', 'Menu', 'Final', 'Taste', 'Quantity', 'Delivery', 'Final P', 'Taste P', 'Quan P', 'Delivery P']
```

```python
df_menu_num_1
```

```python
# 일단 수량 고려하지 않고 계산

df_new = df_menu_num_1[['Menu', 'Final P']]
menu_satis = df_new[df_new['Final P']==1]
menu_satis_count = menu_satis['Menu'].value_counts().to_frame()
menu_satis_count.rename(columns={'Menu':'satis'}, inplace=True)
df_merge = pd.merge(df_menu_count, menu_satis_count, left_index=True, right_index=True, how='outer')
df_merge = df_merge.fillna(0)
```

```python
df_menu_num_1.shape[0]
```

```python
#@title 전체 평점에 대한 확률 계산값

total = df_menu_num_1.shape[0]
satis_num = menu_satis.shape[0] # 만족한 개수
df_merge['menu_order_per'] = df_merge['Menu'] / total
df_merge['satis_per'] = df_merge['satis'] / total
df_merge['satis_menu_per'] = df_merge['satis_per'] / df_merge['menu_order_per']
# df_merge['satis_menu_per_new'] = (df_merge['satis'] / satis_num) * (satis_num / total) / df_merge['menu_order_per'] # 식 검산 
df_merge.sort_values(by='Menu' ,ascending=False).head(20) # 정렬
```

```python
fianl_num = df_menu_num_1[df_menu_num_1['Final P']==1].shape[0]
taste_num = df_menu_num_1[df_menu_num_1['Taste P']==1].shape[0]
quan_num = df_menu_num_1[df_menu_num_1['Quan P']==1].shape[0]
deli_num = df_menu_num_1[df_menu_num_1['Delivery P']==1].shape[0]


fianl_satis = df_menu_num_1[df_menu_num_1['Final P']==1]
taste_satis = df_menu_num_1[df_menu_num_1['Taste P']==1]
quan_satis = df_menu_num_1[df_menu_num_1['Quan P']==1]
deli_satis = df_menu_num_1[df_menu_num_1['Delivery P']==1]

fianl_satis_count = fianl_satis['Menu'].value_counts().to_frame()
taste_satis_count = taste_satis['Menu'].value_counts().to_frame()
quan_satis_count = quan_satis['Menu'].value_counts().to_frame()
deli_satis_count = deli_satis['Menu'].value_counts().to_frame()

fianl_satis_count.rename(columns={'Menu':'final'}, inplace=True)
taste_satis_count.rename(columns={'Menu':'taste'}, inplace=True)
quan_satis_count.rename(columns={'Menu':'quan'}, inplace=True)
deli_satis_count.rename(columns={'Menu':'deli'}, inplace=True)


lst = [df_menu_count, fianl_satis_count, taste_satis_count, quan_satis_count, deli_satis_count] 
lst_merge = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='outer'), lst).fillna(0)

lst_merge['final_per'] = lst_merge['final'] / lst_merge['Menu']
lst_merge['taste_per'] = lst_merge['taste'] / lst_merge['Menu']
lst_merge['quan_per'] = lst_merge['quan'] / lst_merge['Menu']
lst_merge['deli_per'] = lst_merge['deli'] / lst_merge['Menu']
```

```python
#@title 부분 평점에 대한 확률 계산값

lst_merge.sort_values(by='Menu' ,ascending=False).round(2).head(20)
```
